```python
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import ByteLevel
from tokenizers.decoders import ByteLevel as ByteLevelDecoder
from tokenizers.normalizers import NFKC, Sequence

tokenizer = Tokenizer(BPE(unk_token="<UNK>"))
tokenizer.normalizer = Sequence([NFKC()])
tokenizer.pre_tokenizer = ByteLevel(add_prefix_space=True)
tokenizer.decoder = ByteLevelDecoder()

trainer = BpeTrainer(vocab_size=4000, special_tokens=["<PAD>", "<BOS>", "<EOS>"])
tokenizer.train(["TwitterLowerAsciiCorpus.txt"], trainer)

print(tokenizer.encode("you'll be on a whole new level of greatness and you'll probably be tied with beer on joes \"love scales\" if you like biggie").tokens)
tokenizer.save("tokenizer.json")

tokenizer.get_vocab_size()
print(tokenizer.token_to_id('<BOS>'))
print(tokenizer.token_to_id('<EOS>'))
tokenizer.token_to_id('<PAD>')
```

```python
ids = tokenizer.encode("sweetness").ids
tokens = [tokenizer.decode([i], skip_special_tokens=False) for i in ids]
print(tokens)
```
